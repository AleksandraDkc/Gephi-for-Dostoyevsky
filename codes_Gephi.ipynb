```python
import PyPDF2
from itertools import combinations

def extract_text_from_pdf(pdf_path):
    text_per_page = []
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        num_pages = len(reader.pages)
        for page_num in range(num_pages):
            page_text = reader.pages[page_num].extract_text()
            text_per_page.append(page_text)
    return text_per_page

def creer_persos_dico():
    persos_dico = {
        "Rodion Romanovitch Raskolnikov": ["Rodia", "Rodienka", "Rodka"],
        "Poulkhéria Alexandrovna Raskolnikova": [],
        "Evdokia Romanovna": ["Avdotia", "Dounia", "Dounétchka"],
        "Arkadi Ivanovitch Svidrigaïlov": [],
        "Marfa Pètrovna Svidrigaïlovna": [],
        "Filka": [],
        "Piotr Pètrovitch Loujine": [],
        "Sémione Zacharovitch Marméladov": [],
        "Katerina Ivanovna": [],
        "Sonia Sémionovna Marméladovna": ["Sophia", "Sonètchka"],
        "Polenka": ["Polia", "Polètchka"],
        "Lidia": ["Lida", "Lidotchka"],
        "Kolia": ["Kolka"],
        "Praskovia Pavlovna Zarnitsina": ["Pacha", "Pachenka"],
        "Nastassia Pètrovna": ["Nastenka", "Nastassiouchka"],
        "Amalia Fedorovna Lippewechsel": ["Amalia Ludwigovna", "Amalia Ivanovna"],
        "Andreï Sèmionovitch Lébéziatnikov": [],
        "Alona Ivanovna": [],
        "Lisaveta Ivanovna": [],
        "Nicolaï Dèmèntiev": ["Mikolaï", "Nikolka"],
        "Dmitri": ["Mitreï", "Mitka"],
        "Dmitri Prokofitch Vrasoumikhine": ["Rasoumikhine"],
        "Zossimov": [],
        "Guertrouda Karlovna Resslich": [],
        "Kapernaoumov": [],
        "Porfiri Pètrovitch": ["Porphyre"],
        "Ilia Pètrovitch": ["La Poudre"],
        "Nikodim Fomitch": [],
        "Alexandre Grigorievitch Zamètov": []
    }
    return persos_dico

def extract_characters_from_page(page_text):
    persos_dico = creer_persos_dico()
    perso_match = set()  # Utilisation d'un ensemble pour éviter les doublons
    for perso, denominations in persos_dico.items():
        for name in [perso] + denominations:
            if name.lower() in page_text.lower():  # Recherche insensible à la casse
                perso_match.add(perso)
                break
    return list(perso_match)

def count_character_pairs(pages_avec_perso):
    perso_paires_count = {}
    for page_num, perso_in_page in pages_avec_perso:
        for paire in combinations(perso_in_page, 2):
            if paire not in perso_paires_count:
                perso_paires_count[paire] = 0
            perso_paires_count[paire] += 1
    return perso_paires_count

def main():
    pdf_path = "Dostoievski-crime-1.pdf"
    text_per_page = extract_text_from_pdf(pdf_path)
    
    pages_avec_perso = []
    for page_num, page_text in enumerate(text_per_page, start=1):
        perso_in_page = extract_characters_from_page(page_text)
        if perso_in_page:
            pages_avec_perso.append((page_num, perso_in_page))
    
    perso_paires_count = count_character_pairs(pages_avec_perso)
    
    total_relations = sum(perso_paires_count.values())
    total_paires = len(perso_paires_count)
    
    print("Nombre total d'apparitions de paires de personnages par page :")
    for paire, count in perso_paires_count.items():
        print(f"{paire}: {count}")
    print(f"\nNombre total de relations trouvées : {total_relations}")
    print(f"Nombre total de paires trouvées : {total_paires}")

if __name__ == "__main__":
    main()
```
